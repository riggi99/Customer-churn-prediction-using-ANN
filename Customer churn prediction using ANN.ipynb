```python

df_class_0=df1[df1['Churn']==0]
df_class_1=df1[df1['Churn']==1]

df_class_0.shape
df_class_1.shape

df_class_0_New=df_class_0.sample(len(df_class_1))


df_New =pd.concat([df_class_0_New,df_class_1],axis=0)
```

```python
df_New['Churn'].value_counts()
```

```python
x=df_New.drop('Churn', axis = 'columns')
y=df_New.Churn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.2,random_state =0)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([keras.layers.Dense(25, input_shape=(27,), activation='relu'),
                          keras.layers.Dropout(0.2),
                          keras.layers.Dense(15,activation='relu'),
                          keras.layers.Dropout(0.12),
                          keras.layers.Dense(1,activation='sigmoid') 
])
```

```python
model.compile(optimizer='SGD',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test,y_test)
y_pred = model.predict(x_test)
y_pred1=[]
for y in y_pred:
    if y > 0.5:
        y_pred1.append(1)
    else:
        y_pred1.append(0)
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred1))
```
